```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid.csv')

#datatypes
df.dtypes
```

```python
df.head()
```

```python
df.tail()
```

```python
#values on x&y-axis

x = df['date']
y = df['Cases']

plt.title('covid 19 data of Uzbekistan', fontsize=20)
plt.xlabel('years')
plt.ylabel('new cases')


#resize plot
plt.rcParams["figure.figsize"] = (20,7)

plt.plot(x, y, label = 'daily new cases')
plt.legend()
plt.grid()

plt.show()
```

```python
x = df ['date']
y = df ['Death_Cases']

plt.title('covid 19 data of Uzbekistan', fontsize=20 )
plt.xlabel('years')
plt.ylabel('death cases')

#resizing plot
figsize=(8, 6)



plt.scatter(x, y, c = 'red', label = "daily death cases")
plt.legend()
plt.grid(color = 'green', linestyle = '--', linewidth = 0.2)
plt.show()
```

```python

#plot 1:
x = df['date']
y = df['Cases']

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Cases", fontsize=20)


#plot 2:
x = df['date']
y = df['Death_Cases']

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Death Cases", fontsize=20)

plt.suptitle('Ratio between "cases" and "deaths"', fontsize=20)
plt.show()
```

```python
#x = df['date']
y = df['Recovered_Cases']

plt.xlabel('days from 2020-03-15 to 2021-08-01', fontsize=20)
plt.ylabel('recovered cases', fontsize=15)

plt.stem(y)
plt.grid(axis = 'y')
plt.show()
```

```python
#make data
y1 = df["Recovered_Cases"]
y2 = df["Recovered"]


plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()

ax.plot(y1, label = "daily recovered cases" )
ax.plot(y2, label = "total recovered cases")

plt.xlabel('days from 2020-03-15 to 2021-08-01', fontsize=10)
plt.ylabel('recovered cases', fontsize=15)
plt.legend()
plt.show()
```
